🌺 About the Cloth
Material: Tapa cloth was made from the inner bark of trees such as the paper mulberry (Broussonetia papyrifera). The bark was soaked, beaten, and felted into sheets.

Designs: Patterns were applied using natural dyes, stencils, or freehand painting. Motifs often included geometric shapes, leaves, and symbolic imagery tied to genealogy or cosmology.

Function: These textiles were used as clothing, bedding, ceremonial offerings, and even as currency in social exchanges.

The sample you referenced from Alexander Shaw’s Catalogue of the Different Specimens of Cloth (1787) is one of the earliest European attempts to preserve and classify Oceanic material culture. Each copy of Shaw’s book contained actual cloth specimens pasted in, making it both a catalogue and a tactile archive.

🌊 What Captain Cook Might Have Seen
During his voyages, Cook and his crew encountered societies where tapa cloth was woven into the fabric of life:

Tahiti & Society Islands: Women produced tapa in communal settings. Cook observed its role in gift-giving, where chiefs exchanged cloth to cement alliances.

Tonga & Fiji: Large ceremonial cloths (sometimes hundreds of feet long) were presented at feasts, weddings, and funerals, symbolizing wealth and prestige.

Hawaii: Tapa was used for clothing and sacred rituals. Cook’s crew noted the intricate dyeing techniques and the association of certain colors with rank.

New Zealand (Māori): While tapa was less common, Cook encountered woven flax textiles with similar symbolic importance.

Marquesas & Easter Island: Distinct regional styles of barkcloth reflected local aesthetics and spiritual beliefs.

For Cook, these cloths were not only exotic artifacts but windows into the social systems of the Pacific. He saw how textiles embodied status, spirituality, and community identity, far beyond their practical use.
