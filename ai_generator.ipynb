# Rewoven — Prompt Log

**Object:** A Catalogue of the Different Specimens of Cloth collected in the three voyages of Captain Cook to the Southern Hemisphere

**Source:** University of Glasgow Archives and Special Collections

**AI Tool Used:** Dall-E 2

**Role of AI:** Image generator

All prompts submitted to DALL-E 2 during this project, including initial versions, DALL-E's output, and refined follow-up prompts.

---

**My prompt:**
<br><img src="media/Sample1.jpg" width="300">

**DALL-E's output:**
<br><img src="media/Output 1.1.png" width="500">

**Follow-up prompt (after reviewing output):**

**DALL-E's output:**
<br><img src="media/Output 1.2.png" width="500">

**Thoughts:**

However, the second result was not much different even after specifying that this was a book.
